# Planos de color y recorte de una imagen

# PLANOS RGB

```python
#Librerias
import numpy as np
import matplotlib.pyplot as plt
import skimage
from skimage import io
import cv2 as cv
```

```python
#Leer una imagen
Colombia = io.imread('Colombia.jpg')
```

```python
#Visualizacin de la imagen
plt.imshow(Colombia)
```

```python
#Cargamos los diferentes planos de la imagen R-G-B
R_Colombia = Colombia[:,:,2]  #uint8
G_Colombia = Colombia[:,:,1]  #uint8
B_Colombia = Colombia[:,:,0]  #uint8
```

```python
fig, axes = plt.subplots(1, 3,
                         figsize=(22,21),constrained_layout=True)

# Plot 
im = axes[0].imshow(R_Colombia, cmap='gray')
axes[0].axis('off')
axes[0].set_title('R', fontsize=32)

axes[1].imshow(G_Colombia, cmap='gray')
axes[1].axis('off')
axes[1].set_title('G', fontsize=32)

axes[2].imshow(B_Colombia, cmap='gray')
axes[2].axis('off')
axes[2].set_title('B', fontsize=32)
plt.show()
```

# PLANOS HSV

```python
# Convertir imagen RGB to HSV
imgtoHSV = cv.cvtColor(Colombia, cv.COLOR_RGB2HSV)
plt.imshow(imgtoHSV)

H_Colombia = imgtoHSV[:,:,0]  #uint8
S_Colombia = imgtoHSV[:,:,1]  #uint8
V_Colombia = imgtoHSV[:,:,2]  #uint8
```

```python
fig, axes = plt.subplots(1, 3,
                         figsize=(22,21),constrained_layout=True)

# Plot 
im = axes[0].imshow(H_Colombia, cmap='gray')
axes[0].axis('off')
axes[0].set_title('H', fontsize=32)

axes[1].imshow(S_Colombia, cmap='gray')
axes[1].axis('off')
axes[1].set_title('S', fontsize=32)

axes[2].imshow(V_Colombia, cmap='gray')
axes[2].axis('off')
axes[2].set_title('V', fontsize=32)
plt.show()
```
