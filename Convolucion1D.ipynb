```python
#Librerias
import numpy as np
import matplotlib.pyplot as plt
import skimage
from skimage import io
import cv2 as cv
from numpy.fft import fft,ifft
import matplotlib.pyplot as plt
```

```python
#Leer una imagen
Im = io.imread('Colombia.jpg')
plt.imshow(Im)
```

```python
R_Im = Im[:,:,2]  #uint8
G_Im = Im[:,:,1]  #uint8
B_Im = Im[:,:,0]  #uint8
```

```python
#Perfil de linea
pl = R_Im[:, 180]
pl.shape
```

```python
k=np.array([1, 1, 1, 1,1]) #funcion g - conocida
conv = np.convolve(pl,k)    #funcion f - desonocida
```

```python
fig, axes = plt.subplots(3, 1,
                         figsize=(22,21),constrained_layout=True)

# Plot 
im = axes[0].imshow(R_Im, cmap='gray')
axes[0].axis('off')
axes[0].set_title('R', fontsize=32)

axes[1].plot(pl)
axes[1].axis('off')
axes[1].set_title('Perfl de lnea', fontsize=32)

axes[2].plot(conv)
axes[2].axis('off')
axes[2].set_title('Perfl de linea - convolucin', fontsize=32)
plt.show()
```
