# Operaciones de punto -- Procesamiento de im√°genes

```python
#Librerias
import numpy as np
import skimage
from skimage import io
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
#Leer una imagen
Colombia = io.imread('Colombia.jpg')
```

```python
#Visualizaci√≥n de la imagen
plt.imshow(Colombia)
```

```python
b = np.full((Colombia.shape), 100, dtype=None)
#Si le sumamos a la imagen se ve como se pierden los niveles de gris
a = Colombia + b
plt.imshow(a)
```

```python
#Si le restamos, se oscurece la imagen
c = Colombia - b
plt.imshow(c)
```

```python
#Los valores se salen del contenedor, es por ellos que recurrimos a operaciones de Normalizaci√≥n
mul = Colombia * 2
plt.imshow(mul)
```

# Operaciones de punto --- Normalizaci√≥n

```python
#Cargamos los diferentes planos de la imageny visualizamos el plano rojo, posteriormente elegimos solo ese plano para operar
R_Colombia = Colombia[:,:,2]  #uint8
G_Colombia = Colombia[:,:,1]  #uint8
B_Colombia = Colombia[:,:,0]  #uint8

plt.imshow(R_Colombia, cmap='gray')
```

```python
#pasamos a float32 para normalizar
R_Colombiaf32 = np.array(R_Colombia, dtype='float32')
#imagen normalizada
R_nor = R_Colombiaf32/255
#Con la imagen normalizada es posible realizar operaciones sin que salgan del contenedor, es decir, se puede multiplicar y dividir
#y operar de diferentes maneras de modo que no hayan valores menores, mayores o decimales que no admite uint8

R_nor_mul = R_nor**0.3
#plt.imshow(R_nor_mul, cmap='gray')

#Denormalizaci√≥n (uint8)
R_uint8 = R_nor_mul *255
R_uint8 = np.array(R_uint8, dtype='uint8')
plt.imshow(R_uint8, cmap='gray')
```
