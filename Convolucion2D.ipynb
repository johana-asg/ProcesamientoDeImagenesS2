# Convoluci√≥n ---- 2D

```python
#Librerias
import numpy as np
import matplotlib.pyplot as plt
import skimage
from skimage import io
import cv2 as cv
from numpy.fft import fft,ifft
import matplotlib.pyplot as plt
```

```python
#Leer una imagen
Im = cv.imread('Rostros.jpg')
plt.imshow(Im)
```

```python
Im = cv.cvtColor(Im, cv.COLOR_RGB2GRAY)
```

```python
#Kernels para realizar la operaci√≥n de convloluci√≥n 2D
kernel=np.array([[1,1,1],[1,1,1], [1,1,1]])/9 # filtro de promedio- eliminando frecuencias altas #funcion g - conocida
kernelgauss = np.array([[1,2,1],[2, 4, 2], [1,2,1]])/16   #filtro gaussiano
kernel2 = np.array([[1,0,-1],[1,0, -1],[1,0, -1]])
```

```python
#Funci√≥n para calcular la convoluci√≥n
def con2D(Im, kernelx):
    forma = np.shape(Im)
    im2 = np.zeros(forma)
    for x in list(range(1, forma[0]-1)):
        for y in list(range(1, forma[1]-1)):
            suma=0
        
            for i in list(range(-1,2)):
                for j in list(range(-1,2)):
                    suma=Im[x-i, y-j]*kernelx[i+1,j+1]+suma
                im2[x,y]=suma
    maxs = np.max(im2)
    im2 = im2*255/maxs
    im2 = im2.astype(np.uint8)
    return im2
```

```python
im2 = con2D(Im, kernel)
fig, axes = plt.subplots(1, 2,
                         figsize=(22,21),constrained_layout=True)

# Plot 
im = axes[0].imshow(Im, cmap='gray')
axes[0].axis('off')
axes[0].set_title('Imagen original', fontsize=32)

axes[1].imshow(im2, cmap='gray')
axes[1].axis('off')
axes[1].set_title('Imagen convolucionada', fontsize=32)
plt.show()
```

```python
im2 = con2D(Im, kernel2)
fig, axes = plt.subplots(1, 2,
                         figsize=(22,21),constrained_layout=True)

# Plot 
im = axes[0].imshow(Im, cmap='gray')
axes[0].axis('off')
axes[0].set_title('Imagen original', fontsize=32)

axes[1].imshow(im2, cmap='gray')
axes[1].axis('off')
axes[1].set_title('Imagen convolucionada', fontsize=32)
plt.show()
```
