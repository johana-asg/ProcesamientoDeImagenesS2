# Look up Table

```python
#Librerias
import numpy as np
import skimage
from skimage import io
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
#Leer una imagen
Im = io.imread('Colombia.jpg')
plt.imshow(Im)
```

```python
#lookuptable´s gamma
def lut06(image):
    lookUpTable = np.empty((1,256), np.uint8)
    for i in range(256):
        lookUpTable[0,i] = np.clip(pow(i / 255.0, 0.6) * 255.0, 0, 255)
    res = cv.LUT(image, lookUpTable)
    return res

def lut05(image):
    lookUpTable = np.empty((1,256), np.uint8)
    for i in range(256):
        lookUpTable[0,i] = np.clip(pow(i / 255.0, 0.5) * 255.0, 0, 255)
    res = cv.LUT(image, lookUpTable)
    return res


def lut03(image):
    lookUpTable = np.empty((1,256), np.uint8)
    for i in range(256):
        lookUpTable[0,i] = np.clip(pow(i / 255.0, 0.3) * 255.0, 0, 255)
    res = cv.LUT(image, lookUpTable)
    return res

def lut01(image):
    lookUpTable = np.empty((1,256), np.uint8)
    for i in range(256):
        lookUpTable[0,i] = np.clip(pow(i / 255.0, 0.1) * 255.0, 0, 255)
    res = cv.LUT(image, lookUpTable)
    return res

def Negativo(image):
    L = image.max()
    negative = L - image
    return negative
```

# Aplicación LuT

```python
Im_RGB_06 = lut06(Im)
plt.imshow(Im_RGB_06)
```

```python
Im_RGB_05 = lut05(Im)
plt.imshow(Im_RGB_05)
```

```python
Im_RGB_03 = lut03(Im)
plt.imshow(Im_RGB_03)
```

```python
Im_RGB_01 = lut01(Im)
plt.imshow(Im_RGB_01)
```

```python
#Si me interesa una región en particular
Im_RGB_03 = lut01(Im[0:350, 200:650])
plt.imshow(Im_RGB_03)
```

```python
Im_Neg = Negativo(Im)
plt.imshow(Im_Neg)
```
